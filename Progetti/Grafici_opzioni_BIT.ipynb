```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file Excel
df = pd.read_excel("risultati2.xlsx")

# Colonne da visualizzare (supponendo che A sia la prima colonna, cioè la 0-esima)
colonna_x = 0  # Colonna A
colonna_y1 = 7  # Colonna 8
colonna_y2 = 8  # Colonna 9

# Converti le colonne in numerico (se necessario)
df.iloc[:, colonna_y1] = pd.to_numeric(df.iloc[:, colonna_y1], errors='coerce')
df.iloc[:, colonna_y2] = pd.to_numeric(df.iloc[:, colonna_y2], errors='coerce')

# Crea una figura con due sottografi disposti in orizzontale
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 1 grafico: Colonna 8 a partire dalla riga 4
ax1.plot(df.iloc[3:, colonna_x], df.iloc[3:, colonna_y1], label='Call Price media Buy-Sell')
ax1.set_xlabel('Strike Price(K)')
ax1.set_ylabel('Prezzo')
ax1.set_title('Prezzi Borsa IT - Opzioni Call (Enel)')
ax1.legend()
ax1.grid(True)

# 2 grafico: Colonna 9 a partire dalla riga 4
ax2.plot(df.iloc[3:, colonna_x], df.iloc[3:, colonna_y2], label='Put Price media Buy-Sell')
ax2.set_xlabel('Strike Price(K)')
ax2.set_ylabel('Prezzo')
ax2.set_title('Prezzi Borsa IT - Opzioni Put (Enel)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
