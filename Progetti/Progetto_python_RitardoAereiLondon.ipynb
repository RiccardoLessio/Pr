```python
#In ottica di lavoro semplificato a chi deve analizzare il database, creiamo una funzione per ogni comando 
#utile all'analisi dei dati
```

```python
#Creiamo un dataframe, che è ripulito e leggibile, selezionabile con la variabile del nome
#dell'aeroporto che vogliamo analizzare. Di fatto è tutto il codice scritto in precedenza ma generalizzato
#in modo da usarlo per ogni aeroporto di Londra.

def analisi_aeroporto(df):
    df=df.set_index("origin_destination")

    
    
    df=df.groupby("airline_name").mean()
    df.pop('number_flights_cancelled')
    
    df["Voli in anticipo %"]=df["flights_more_than_15_minutes_early_percent"]+df["flights_15_minutes_early_to_1_minute_early_percent"]
    del df["flights_more_than_15_minutes_early_percent"]
    del df["flights_15_minutes_early_to_1_minute_early_percent"]
    
    df["Voli in ritardo 0-30 minuti %"]=df["flights_0_to_15_minutes_late_percent"]+df["flights_between_16_and_30_minutes_late_percent"]
    del df["flights_0_to_15_minutes_late_percent"]
    del df["flights_between_16_and_30_minutes_late_percent"]
    
    df["Voli in ritardo 31-120 minuti %"]=df["flights_between_31_and_60_minutes_late_percent"]+df["flights_between_61_and_120_minutes_late_percent"]
    del df["flights_between_31_and_60_minutes_late_percent"]
    del df["flights_between_61_and_120_minutes_late_percent"]
    
    df["Voli con più di 120 minuti di ritardo %"]=df["flights_between_121_and_180_minutes_late_percent"]+df["flights_between_181_and_360_minutes_late_percent"]+df["flights_more_than_360_minutes_late_percent"]
    del df['flights_between_121_and_180_minutes_late_percent']
    del df['flights_between_181_and_360_minutes_late_percent']
    del df['flights_more_than_360_minutes_late_percent']
    df=df.loc[~(df==0).all(axis=1)] 
    return df

#Il return alla fine della funzione è necessario sia per visualizzzare il dataframe
#che per importare i dati nelle successive funzioni

def var_random(df):
    var_random=random.randint(0,10000)
    random_aircomp=df.sample(n=10, random_state=var_random)
    return random_aircomp
    
def grafico_completo(df):
    grafico_barre=df.plot.bar(figsize=(30,10))
    grafico_barre.set_ylabel("Percentuale")
    grafico_barre.set_xlabel("Compagnia Aerea")
    grafico_barre.set_title("Analisi anticipi/ritardi per compagnia aerea")
    grafico_barre.legend(title="Legenda")
    return grafico_barre
    
def grafico_random(df):    
    grafico_barre_random=df.plot.bar(figsize=(20,10))
    grafico_barre_random.set_ylabel("Percentuale")
    grafico_barre_random.set_xlabel("Compagnia Aerea")
    grafico_barre_random.set_title("Analisi anticipi/ritardi per compagnia aerea casuale")
    grafico_barre_random.legend(title="Legenda")
    #abbiamo provato a ruotare di 45° i nominativi delle compagnie aeree sull'asse x con la linea di codice 
    #"grafico_barre_random.set_xticklabels(df.index, rotation=45)" ma i nomi troppo lunghi sfasavano l'inetro asse
    return grafico_barre_random
```

```python
#Per visualizzare le funzioni dobbiamo salvarle in una variabile e poi richiamarla
#Così facendo scrivendo tra parentesi la variabile contenente il dataframe dell'aeroporto, la funzione
#ripulisce il dataframe stesso e lo restituisce con i dati più importanti.
data = analisi_aeroporto(London_HTR)
data
```

```python
#dando in input il dataframe generato nella funzione "analisi_aeroporto", salvato in "data", 
#fornisce il dataframe randomico
```

```python
var = var_random(data)
var
```

```python
#In base all'aeroporto analizzato restituisce il grafico "completo" con tutte le compagnie aeree.
grafico_completo(data)
```

```python
#Questo è il grafico randomico fornendo in input il dataframe calcolato col metodo random
grafico_random(var)
```
