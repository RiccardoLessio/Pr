```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
import random


#font
font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
font.set_style('italic')
#stileplot
#plt.style.use('seaborn-v0_8-notebook')
#plt.style.use('default')
plt.style.use('seaborn-v0_8-whitegrid')
#titolo
plt.suptitle("Random walk e profitto",fontsize=14,style='italic')#,weight='bold'

#nomi assi
plt.xlabel("Tempo t", fontsize=12,style='italic')
plt.ylabel("Prezzo St", fontsize=12,style='italic')

#colore nuemero assi
plt.xticks(color="black") #w=white b=blu
plt.yticks(color="black") #w=white b=blu

#griglia
plt.grid(True)


def randomwalk1D(n,x,y):

    # Generate the time points [1, 2, 3, ... , n]
    timepoints = np.arange(n + 1)
    positions = [y]
    directions = ["UP", "DOWN"]
    for i in range(1, n + 1):
        # Randomly select either UP or DOWN
        step = random.choice(directions)

        # Move the object up or down
        if step == "UP":
            y += .01*y
        elif step == "DOWN":
            y -= .012*y
        # Keep track of the positions
        positions.append(y)
    return timepoints, positions
rw1 = randomwalk1D(1000,0,1.8)
rw2 = randomwalk1D(1000,0,1)

plt.plot(rw1[0], rw1[1], 'black', label="rw1")

```
