```python
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties
#size
plt.rcParams["figure.figsize"] = (8, 5)

#font
font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
font.set_style('italic')
#stileplot
#plt.style.use('seaborn-v0_8-notebook')
#plt.style.use('default')
plt.style.use('seaborn-v0_8-whitegrid')
#titolo
plt.suptitle("Payoff opzione call (Short Position)",fontsize=14,style='italic')#,weight='bold'

#nomi assi
plt.xlabel("Prezzo Sottostante", fontsize=12,style='italic')
plt.ylabel("Profitto", fontsize=12,style='italic')

#colore nuemero assi
plt.xticks(color="black") #w=white b=blu
plt.yticks(color="black") #w=white b=blu

#griglia
plt.grid(True)

#lunghezza assi
plt.axis ((-5,20,-20,20))

#figure
plt.plot([10],[0],"go", color="black") #punto1
plt.plot([8.25],[+3],"go", color="black") #punto2
plt.plot([19.8],[0],">", color="black") #freccia
plt.plot([0],[19.5],"^", color="black") #freccia
plt.plot([0,8.25,10,30],[+3,+3,0,-35], color="black", linewidth=2)#linea
plt.plot([0,0],[-20,20], color="black", linewidth=1.07)#linea
plt.plot([-5,20],[0,0], color="black", linewidth=1.07)#linea
plt.text(8, +4.5, r'K',fontsize=16,style='italic')#testo
plt.text(9.5, 1.5, r'BEP',fontsize=14,style='italic')#testo

plt.text(0.4, 18, r'Ï€',fontsize=16,style='italic')#testo
plt.text(19, 0.9, r'St',fontsize=16,style='italic')#testo
plt.text(-0.7, +2, r'c',fontsize=16,style='italic')#testo
plt.show()
```
